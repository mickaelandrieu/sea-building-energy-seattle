The `Zip Codes` variable from 2015 data looks invalid, we will rebuild it from the `Location` variable.

```python
import ast

def extract_data_from_location(row: pd.Series) -> pd.Series:
    building = row.copy()
    parsed_location = ast.literal_eval(building.Location)
    building['Latitude'] = parsed_location['latitude']
    building['Longitude'] = parsed_location['longitude']
    parsed_human_address = ast.literal_eval(parsed_location['human_address'])
    building['Address'] = parsed_human_address['address']
    building['City'] = parsed_human_address['city']
    building['State'] = parsed_human_address['state']
    building['ZipCode'] = parsed_human_address['zip']
    
    return building

df_2015 = df_2015.apply(extract_data_from_location, axis='columns')

df_2015.head()
```

```python
df_2015.rename(columns={
    'Comment': 'Comments',
    'GHGEmissions(MetricTonsCO2e)': 'TotalGHGEmissions',
    'GHGEmissionsIntensity(kgCO2e/ft2)': 'GHGEmissionsIntensity'
}, inplace=True)

df_2015
```

The targets to predict are the following:
* `TotalGHGEmissions`
* `Electricity(kBtu)`

> Btu stands for "British thermal unit" : 1kWh = 3.412 kBtu (direct relationship)

They chose this target over `Electricity(kWh)` because most of the other related variables are also in Btu.

Now let's clean some variables names and suffix our targets with `target` so we remember which variables we want to predict.

```python
buildings.columns
```

```python
def clean_variables_names(df):
    df.columns = df.columns.str.strip().str.replace(r'(?<!^)(?=[A-Z])', '_', regex=True).str.lower()
    
    df.rename(columns={
        'o_s_e_building_i_d': 'building_id',
        'numberof_buildings': 'number_of_buildings',
        'numberof_floors': 'number_of_floors',
        'property_g_f_a_total': 'property_gfa_total',
        'property_g_f_a_parking': 'property_gfa_parking',
        'property_g_f_a_building(s)': 'property_gfa_building',
        'largest_property_use_type_g_f_a': 'largest_property_use_type_gfa',
        'second_largest_property_use_type_g_f_a': 'second_largest_property_use_type_gfa',
        'third_largest_property_use_type_g_f_a': 'third_largest_property_use_type_gfa',
        'years_e_n_e_r_g_y_s_t_a_r_certified': 'years_energy_star_certified',
        'e_n_e_r_g_y_s_t_a_r_score': 'energystar_score',
        'site_e_u_i(k_btu/sf)': 'site_eui',
        'site_e_u_i_w_n(k_btu/sf)': 'site_euiwn',
        'source_e_u_i(k_btu/sf)': 'source_eui',
        'source_e_u_i_w_n(k_btu/sf)': 'source_euiwn',
        'site_energy_use(k_btu)': 'site_energy_use',
        'site_energy_use_w_n(k_btu)': 'site_energy_use_wn',
        'steam_use(k_btu)': 'steam_use',
        'electricity(k_wh)': 'electricity_kwh',
        'electricity(k_btu)': 'electricity_target',
        'natural_gas(therms)': 'natural_gas_therms',
        'natural_gas(k_btu)': 'natural_gas',
        'total_g_h_g_emissions': 'emissions_target',
        'g_h_g_emissions_intensity': 'emissions_intensity'
    }, inplace=True)
    
    return df

clean_variables_names(buildings)

buildings.columns
```

```python
buildings
```

We think the variable `building_id` is an unique identifier, but we only need to keep one entry for each unique entry.

This means we must agregate every numerical variable while grouping on the `building_id`:

```python
# @TODO create the corresponding function
numerical_variables = list(buildings.select_dtypes('number'))
unique_buildings = buildings[numerical_variables].groupby('building_id').mean()
unique_buildings['is_agregation'] = unique_buildings['data_year'] == 2015.5
unique_buildings
```

```python
# now let's keep only the last value when there are duplications
deduplicated_buildings = buildings.sort_values('data_year').drop_duplicates(subset = ['building_id'], keep = 'last')

deduplicated_buildings
```

Now we need to merge the categorical and numerical variables :

```python
numerical_variables.remove('building_id')

deduplicated_buildings.drop(numerical_variables, axis = 'columns', inplace = True)
final_buildings = deduplicated_buildings.merge(unique_buildings, on = 'building_id', how = 'left')
final_buildings
```
